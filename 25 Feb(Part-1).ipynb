#### Groupby function : Bring thw similar data together

```python
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.groupby('Survived') # It gives the object 
```

```python
# What is the average fare paid by people who died and people who survived
df.groupby('Survived').mean()
```

```python
# WHat is the sum of  fare paid by people who died and people who survived
df.groupby('Survived').sum()
```

```python
 ## What is the total revenue paid by women and what renueue paid by men
df.groupby('Sex').sum()['Fare']
```

```python
 ## What is the average age of women and men
df.groupby('Sex').mean()['Age']
```

```python
## What average booking done by class 1,class 2 ,class 3 passengers
df.groupby('Pclass').mean()['Fare']
```

#### Merge operation (INTERSECTION)

```python
pd.merge(data1,data2)   
# In merge operation it will display only the common values of both datasets
# In this if there is a common column then print the all common possible 
```

```python
pd.merge(data1,data2,how='left') 
# data1 is left data2 is right (left is complete whatever match in right write correspond value o/w NaN)
```

```python
pd.merge(data1,data2,how='right') 
# data1 is left data2 is right (right is complete whatever match in left write correspond value o/w NaN)
```

```python
pd.merge(data1,data2,how='inner')   #The by-default merge operation is inner
# It takes common from both left and right
```

```python
pd.merge(data1,data2,how='outer') 
# It takes everything from both left and right and put NaN if not presnt in the other dataset
```

```python
pd.merge(data1,data2,how='cross') 
# It takes every data from first and merge with everything of the second 
```

#### Join operation  :
Difference b/w merge and join

In merge it tries to match columnwise while in join it tries to match indices(row)

```python
#Create 2 datasets with their own custom indices
```

```python
dataj1=pd.DataFrame({ "key1":[1,2,4,5,6],
                     "key2":[4,5,6,7,8],
                     "key3":[3,4,5,6,6]},
                    index=['a','b','c','d','e'])
```

```python
dataj1
```

```python
dataj2=pd.DataFrame({ "key11":[1,2,4,5,6],
                     "key22":[4,5,6,7,8],
                     "key33":[3,4,5,6,6]},
                    index=['a','b','h','i','j'])
```

```python
dataj2
```

```python
dataj1.join(dataj2)   #Bydefault join by left
```

```python
dataj1.join(dataj2,how='inner')  #Givee only the common values
```

```python
dataj1.join(dataj2,how='outer')  #Take every values from both datasets and place NaN accordingly
```

```python
dataj1.join(dataj2,how='right')  # Check with dataj2 
```

```python
dataj1.join(dataj2,how='cross')  #Give all possible joins
```
